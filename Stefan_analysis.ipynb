```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp 
```

```python
xstar = lambda lmbda,t : 2*lmbda*np.sqrt(t)
lfunc = lambda lmbda : ste*lmbda - 1/np.sqrt(np.pi) * (np.exp(-lmbda**2))/sp.special.erf(lmbda)
T = lambda x ,t : 1 - sp.special.erf(x/(2*np.sqrt(t)))/sp.special.erf(lmbda)
```

```python
Tmin , Tmax = 300, 3000
cps , cpl = 750, 1000
lheat = 4e5
Tsol, Tliq = 1475, 1525
ste = (cps*(Tsol - Tmin) + cpl*(Tmax - Tliq))/lheat
```

```python
from scipy.optimize import fsolve
```

```python
fsolve(lfunc, 1)
```
